```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
# from BBH_60_30_mass import *
from BBH_100_20_mass import *
```

**Creat Data**

```python
sample_rate = 1024
training_data, M1, M2 = bbhinspiral(sample_rate, 30000)
testing_data, t_M1, t_M2 = bbhinspiral(sample_rate, 5000)
validation_data, v_M1, v_M2 = bbhinspiral(sample_rate, 15000)
```

**Save Data**

```python
path = 'Data_20_100.h5'
with h5py.File(path, 'w') as ST:
    g1 = ST.create_group('Training_data')
    g1.create_dataset('Strain', data = training_data)
    g1.create_dataset('Mass_1', data = M1)
    g1.create_dataset('Mass_2', data = M2)
    g2 = ST.create_group('Testing_data')
    g2.create_dataset('Strain', data = testing_data)
    g2.create_dataset('Mass_1', data = t_M1)
    g2.create_dataset('Mass_2', data = t_M2)
    g3 = ST.create_group('Validation_data')
    g3.create_dataset('Strain', data = validation_data)
    g3.create_dataset('Mass_1', data = v_M1)
    g3.create_dataset('Mass_2', data = v_M2)
```

**Data Info**

```python
def display(name):
    print(name)
# path = 'Data.h5'
Data = h5py.File(path, 'r')
Data.visit(display)
```
