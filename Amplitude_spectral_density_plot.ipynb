```python
# Info

# This file shows the plot of amplitude spectral density of noise versus it's frequentz
#     measured at Livingston and Hanford.

# Reference : https://www.gw-openscience.org/events/GW150914/

# Data source L1 : https://www.gw-openscience.org/GW150914data/P1500238/L1-GDS-CALIB_STRAIN.txt
#             H1 : https://www.gw-openscience.org/GW150914data/P1500238/H1-GDS-CALIB_STRAIN.txt
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L1_Strain = np.loadtxt("/Users/acye/Documents/Outrun/Sparrow/Program/Copycat/Data/L1-GDS-CALIB_STRAIN.txt")
H1_Strain = np.loadtxt("/Users/acye/Documents/Outrun/Sparrow/Program/Copycat/Data/H1-GDS-CALIB_STRAIN.txt")
```

```python
x1, y1 = L1_Strain[:,0], L1_Strain[:,1]
x2, y2 = H1_Strain[:,0], H1_Strain[:,1] 
```

```python
plt.figure(dpi = 500)
plt.grid()
plt.xlabel('Frequency [Hz]')
plt.ylabel('ASD h(f)')
plt.ylim(10**-24.5,10**-19)
plt.xlim(10,5000)
plt.loglog(x1,y1, label = 'L1')
plt.loglog(x2,y2, label = 'H2')
plt.legend()
plt.show()
```
